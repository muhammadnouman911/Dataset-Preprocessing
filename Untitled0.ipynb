```python
def download_video(video_id):
    output_path = f"videos/{video_id}.mp4"
    if os.path.exists(output_path):
        return output_path
    url = f"https://www.youtube.com/watch?v={video_id}"
    os.makedirs("videos", exist_ok=True)
    try:
        !yt-dlp -f mp4 -o "videos/{video_id}.mp4" {url}
        return output_path if os.path.exists(output_path) else None
    except Exception as e:
        print(f"Download error: {e}")
        return None
```

```python
def extract_clip(video_path, start, end, output_path):
    try:
        clip = VideoFileClip(video_path).subclip(max(0, float(start)-1), float(end)+1)
        clip.write_videofile(output_path, codec="libx264", audio_codec="aac", verbose=False, logger=None)
        return output_path
    except Exception as e:
        print(f"❌ Clip extraction error: {e}")
        return None
```

```python
def extract_audio(video_path, output_audio):
    try:
        clip = VideoFileClip(video_path)
        clip.audio.write_audiofile(output_audio, verbose=False, logger=None)
    except Exception as e:
        print(f"❌ Audio error: {e}")
```

```python
def calculate_snr(audio_path):
    try:
        y, sr = librosa.load(audio_path, sr=None)
        signal_power = np.mean(y**2)
        noise_power = np.var(y - np.mean(y))
        if noise_power == 0:
            return 100
        snr = 10 * np.log10(signal_power / noise_power)
        return snr
    except:
        return 0
```

```python
def extract_face_crop(video_path, output_image):
    try:
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        cap = cv2.VideoCapture(video_path)
        mid_frame = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) // 2
        cap.set(cv2.CAP_PROP_POS_FRAMES, mid_frame)
        ret, frame = cap.read()
        cap.release()

        if not ret:
            return

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)

        if len(faces) > 0:
            x, y, w, h = faces[0]
            face_crop = frame[y:y+h, x:x+w]
            cv2.imwrite(output_image, face_crop)
        else:
            # Save full frame as fallback
            cv2.imwrite(output_image, frame)
    except Exception as e:
        print(f"❌ Face extraction error: {e}")
```

```python
# Load CSV
df = pd.read_csv("data.csv")
os.makedirs("dataset", exist_ok=True)

for idx, row in df.iterrows():
    try:
        yt_id, start, end, x, y = row
        folder = f"dataset/A{idx+1}"
        os.makedirs(folder, exist_ok=True)

        print(f"▶️ Processing A{idx+1}: {yt_id}")

        # Download video
        video_path = download_video(yt_id)
        if not video_path:
            print(f"❌ Video not found: {yt_id}")
            continue

        # Trim video
        segment_path = f"{folder}/segment.mp4"
        trimmed = extract_clip(video_path, start, end, segment_path)
        if not trimmed:
            continue

        # Extract audio
        audio_path = f"{folder}/audio.wav"
        extract_audio(trimmed, audio_path)

        # SNR filter
        snr = calculate_snr(audio_path)
        if snr < 0:
            print(f"⚠️ Low SNR ({snr:.2f} dB), skipping A{idx+1}")
            shutil.rmtree(folder)
            continue

        # Extract face
        face_path = f"{folder}/face.jpg"
        extract_face_crop(trimmed, face_path)

        print(f"✅ Done: A{idx+1}")

    except Exception as e:
        print(f"❌ Error on A{idx+1}: {e}")
```

```python
import os
import uuid
import cv2
import numpy as np
from pytube import YouTube
from moviepy.editor import VideoFileClip
from pydub import AudioSegment
import webrtcvad
import librosa
import soundfile as sf
import pandas as pd

# Directories
VIDEO_DIR = "/content/videos"
DATASET_DIR = "/content/dataset_auto"
os.makedirs(VIDEO_DIR, exist_ok=True)
os.makedirs(DATASET_DIR, exist_ok=True)

# Download YouTube video
def download_video(youtube_id):
    try:
        yt = YouTube(f"https://www.youtube.com/watch?v={bz0hvzo6pMg}")
        stream = yt.streams.filter(file_extension="mp4", progressive=True).order_by("resolution").desc().first()
        path = os.path.join(VIDEO_DIR, f"{youtube_id}.mp4")
        stream.download(output_path=VIDEO_DIR, filename=f"{youtube_id}.mp4")
        return path
    except Exception as e:
        print(f"[❌] Failed to download {youtube_id}: {e}")
        return None

# Detect face
def detect_face(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml").detectMultiScale(gray, 1.3, 5)
    return faces[0] if len(faces) > 0 else None

# Voice Activity Detection
def is_speech(audio_path, sample_rate=16000):
    try:
        audio, sr = librosa.load(audio_path, sr=sample_rate)
        if len(audio) < sample_rate:
            audio = np.pad(audio, (0, sample_rate - len(audio)))
        pcm_data = (audio * 32768).astype(np.int16).tobytes()
        vad = webrtcvad.Vad(2)
        frame_duration = 30
        frame_size = int(sample_rate * frame_duration / 1000) * 2
        num_frames = len(pcm_data) // frame_size
        for i in range(num_frames):
            frame = pcm_data[i * frame_size:(i + 1) * frame_size]
            if vad.is_speech(frame, sample_rate):
                return True
        return False
    except:
        return False

# Process each video
def process_video(youtube_id):
    path = download_video(youtube_id)
    if not path:
        return []

    cap = cv2.VideoCapture(path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    duration = cap.get(cv2.CAP_PROP_FRAME_COUNT) / fps

    sample_every = 5
    current = 0
    collected = []

    while current + sample_every < duration:
        cap.set(cv2.CAP_PROP_POS_MSEC, current * 1000)
        ret, frame = cap.read()
        if not ret:
            break

        face = detect_face(frame)
        if face is not None:
            x, y, w, h = face
            face_img = frame[y:y+h, x:x+w]
            uid = f"A{uuid.uuid4().hex[:4]}"
            folder = os.path.join(DATASET_DIR, uid)
            os.makedirs(folder, exist_ok=True)

            # Save face
            img_path = os.path.join(folder, "face.jpg")
            cv2.imwrite(img_path, face_img)

            try:
                clip = VideoFileClip(path).subclip(current, current + sample_every)
                audio_path = os.path.join(folder, "audio.wav")
                clip.audio.write_audiofile(audio_path, verbose=False, logger=None)

                if not is_speech(audio_path):
                    os.remove(audio_path)
                    os.remove(img_path)
                    os.rmdir(folder)
                else:
                    collected.append({
                        "id": uid,
                        "video": youtube_id,
                        "start": current,
                        "end": current + sample_every,
                        "x": int(x + w/2),
                        "y": int(y + h/2),
                        "folder": folder
                    })
            except Exception as e:
                print(f"[⚠️] Audio extract failed: {e}")
        current += sample_every

    cap.release()
    return collected

# YouTube video IDs
youtube_ids = ["dQw4w9WgXcQ"]  # Replace with your own

# Main
all_clips = []
for yid in youtube_ids:
    clips = process_video(yid)
    all_clips.extend(clips)

# Save metadata
df = pd.DataFrame(all_clips)
df.to_csv("/content/generated_clips_metadata.csv", index=False)
print("✅ Dataset created at: /content/dataset_auto")
```
